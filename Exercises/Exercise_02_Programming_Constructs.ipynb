# Exercise 2.1 Unit Converter

Write a program that helps the user to convert a given quantity from one unit of (distance, area or volume) measurement to another.

The problem:
-	Request a quantity from the user.
-	Have the user specify if the quantity is a `length`, `area` or `volume`.
-	Have the user select the base unit of measurement from the following list:
    - `mm`,
    - `cm`,
    - `m`, or
    - `km`.
-	Then have the user select the desired base unit of measurement that the given quantity should be converted to.
-	Print out converted quantity (including the new unit of measurement).

Example interaction:

>``` text
>Input a quantity: 10
>Select either length, area or volume: area
>Select the base unit of measurement (mm, cm, m, km) : m
>Select which base unit of measurement to convert to (mm, cm, m, km) : cm
>10 m^2 = 100000 cm^2
>```

````python
#YOUR CODE HERE
# Unit Converter
quantity = float(input("Input a quantity: "))
measure_type = input("Select either length, area or volume: ")
from_unit = input("Select the base unit of measurement (mm, cm, m, km) : ")
to_unit = input("Select which base unit of measurement to convert to (mm, cm, m, km) : ")

# Define conversion factors to meters (base unit)
to_meters = {
    'mm': 0.001,
    'cm': 0.01,
    'm': 1,
    'km': 1000
}

# Determine the power based on measure type
if measure_type == 'length':
    power = 1
    unit_suffix = ''
elif measure_type == 'area':
    power = 2
    unit_suffix = '^2'
elif measure_type == 'volume':
    power = 3
    unit_suffix = '^3'

# Convert to base unit (meters) then to target unit
base_quantity = quantity * (to_meters[from_unit] ** power)
converted_quantity = base_quantity / (to_meters[to_unit] ** power)

print(f"{quantity} {from_unit}{unit_suffix} = {converted_quantity} {to_unit}{unit_suffix}")
````

# Exercise 2.2

In mathematics, an **arithmetic sequence** is a sequence of real numbers
$$u_1,u_2,\cdots, u_n$$
such that the difference between two consecutive terms is constant, i.e. $u_{n+1}-u_n=d$ for some constant $d\in \mathbb{R}$, called the **common difference** and $n\geq 1$.

For example, the sequence $2,5,8,11,14,\dots$ is an arithmetic sequence with first term $2$ and common difference $3$.

Write a program to generate the first $n$ terms of an arithmetic sequence with a user defined first term $a$ and common difference $d$.

The problem:
- Request a numeric input from the user to be used as the first term.  Let us denote this input as `first_term`.
- Request a numeric input from the user to be used as the common difference.  Let us denote this input as `common_difference`.
- Request an integer input from the user for the number of terms.  Let us denote this input as `number_of_terms`.
- Print out the sequence up to the $n$th term.

Example interaction:

>``` text
>Please input the first term in the arithmetic sequence:
>2
>Please input the common difference:
>3
>Please input the number of terms:
>5
>The first 5 terms of the arithmetic sequence are:
>2
>5
>8
>11
>14
>```

````python
#YOUR CODE HERE
# Arithmetic Sequence Generator
first_term = float(input("Please input the first term in the arithmetic sequence:\n"))
common_difference = float(input("Please input the common difference:\n"))
number_of_terms = int(input("Please input the number of terms:\n"))

print(f"The first {number_of_terms} terms of the arithmetic sequence are:")
for i in range(number_of_terms):
    term = first_term + i * common_difference
    print(term)
````

# Exercise 2.3

Given an arithmetic sequence with first term $a$ and common difference $d$.

Write a program that:
- Request a numeric input from the user to be used as the first term.  Let us denote this input as `first_term`.
- Request a numeric input from the user to be used as the common difference.  Let us denote this input as `common_difference`.
- Request an integer input from the user for the index for the term in the sequence.  Let us denote this input as `n`.
- Print out the $n$th term in the sequence.
- Print out the sum of the first $n$ terms in the sequence.

Example interaction:

>```text
>Please input the first term in the arithmetic sequence:
>2
>Please input the common difference:
>3
>Please input the index :
>5
>The 5th term in the sequence is:
>14
>The sum of the first 5 terms in the sequence is:
>40
>```

````python
#YOUR CODE HERE
# Arithmetic Sequence nth term and sum
first_term = float(input("Please input the first term in the arithmetic sequence:\n"))
common_difference = float(input("Please input the common difference:\n"))
n = int(input("Please input the index :\n"))

# Calculate nth term: a_n = a + (n-1)d
nth_term = first_term + (n - 1) * common_difference
print(f"The {n}th term in the sequence is:")
print(nth_term)

# Calculate sum of first n terms: S_n = n/2 * (2a + (n-1)d)
sum_n = n / 2 * (2 * first_term + (n - 1) * common_difference)
print(f"The sum of the first {n} terms in the sequence is:")
print(int(sum_n))
````

# Exercise 2.4

An arithmetic sequence $u_1, u_2, u_3, \cdots$ is defined by $u_r=3r-1$ for $r\in \mathbb{Z}^{+}$. Find the least value of $n$ such that $u_3+u_6+u_9+\cdots+u_n$ exceeds $1700$, where $n$ is a multiple of 3.

````python
#YOUR CODE HERE
# Exercise 2.4: Find least n where sum exceeds 1700
# u_r = 3r - 1, we need u_3, u_6, u_9, ... (multiples of 3)
# So u_3 = 3(3) - 1 = 8, u_6 = 3(6) - 1 = 17, u_9 = 3(9) - 1 = 26
# This forms an arithmetic sequence with first term 8 and common difference 9

total_sum = 0
n = 3
while total_sum <= 1700:
    u_n = 3 * n - 1
    total_sum += u_n
    if total_sum > 1700:
        print(f"The least value of n is: {n}")
        print(f"Sum = {total_sum}")
        break
    n += 3
````

On 1 January 2001 Mrs $A$ put \$100 into a bank account, and on the first day of each subsequent month she put in \$10 more than in the previous month. Thus on 1 February she put \$110 into the account and on 1 March she put \$120 into the account, and so on. The account pays no interest.

On what date did the value of Mrs $A$'s account first become greater than $5000?

````python
#YOUR CODE HERE
# Exercise 2.5: Mrs A's account
# January 1, 2001: $100
# February 1, 2001: $110
# March 1, 2001: $120
# Arithmetic sequence: first term = 100, common difference = 10

from datetime import datetime, timedelta

total = 0
month = 0
deposit = 100

start_date = datetime(2001, 1, 1)

while total <= 5000:
    total += deposit
    month += 1
    if total > 5000:
        current_date = start_date + timedelta(days=30*(month-1))
        # More accurate: calculate the actual date
        year = 2001
        month_num = month
        while month_num > 12:
            year += 1
            month_num -= 12
        print(f"The account first exceeded $5000 on month {month}")
        print(f"Date: {month_num}/1/{year}")
        print(f"Total: ${total}")
        break
    deposit += 10
````

## Exercise 2.8

In mathematics, a **geometric sequence** is a sequence of real numbers
$$u_1,u_2,\cdots, u_n$$
such that the ratio between two consecutive terms is constant, i.e. $\frac{u_{n+1}}{u_n}=r$ for some constant $r\in \mathbb{R}\backslash \{0,1\}$, called the **common ratio** and $n\geq 1$.

For example, the sequence $2,6,18,54,162,\dots$ is an geometric sequence with first term $2$ and common ratio $3$.

Write a program to generate the first $n$ terms of an geometric sequence with a user defined first term $a$ and common ratio $r$.

The problem:
- Request a numeric input from the user to be used as the first term.  Let us denote this input as `first_term`.
- Request a numeric input from the user to be used as the common difference.  Let us denote this input as `common_ratio`.

>```text
>Please input the first term in the geometric sequence:
>2
>Please input the common ratio:
>3
>Please input the number of terms:
>5
>The first 5 terms of the geometric sequence are:
>2
>6
>18
>54
>162
>```

## Exercise 2.9

Given a geometric sequence with first term $a$ and common ratio $r$.

Write a program that:
- Request a numeric input from the user to be used as the first term.  Let us denote this input as `first_term`.
- Request a numeric input from the user to be used as the common difference.  Let us denote this input as `common_ratio`.
- Request an integer input from the user for the index for the term in the sequence.  Let us denote this input as `n`.
- Print out the $n$th term in the sequence.
- Print out the sum of the first $n$ terms in the sequence.

Example interaction:

>```text
>Please input the first term in the geometric sequence:
>2
>Please input the common ratio:
>3
>Please input the index :
>5
>The 5th term in the sequence is:
>162
>The sum of the first 5 terms in the sequence is:
>242
>```

````python
#YOUR CODE HERE
# Exercise 2.9: Geometric Sequence nth term and sum
first_term = float(input("Please input the first term in the geometric sequence:\n"))
common_ratio = float(input("Please input the common ratio:\n"))
n = int(input("Please input the index :\n"))

# Calculate nth term: a_n = a * r^(n-1)
nth_term = first_term * (common_ratio ** (n - 1))
print(f"The {n}th term in the sequence is:")
print(int(nth_term) if nth_term == int(nth_term) else nth_term)

# Calculate sum of first n terms: S_n = a(r^n - 1)/(r - 1) for r != 1
if common_ratio == 1:
    sum_n = first_term * n
else:
    sum_n = first_term * (common_ratio ** n - 1) / (common_ratio - 1)
print(f"The sum of the first {n} terms in the sequence is:")
print(int(sum_n) if sum_n == int(sum_n) else sum_n)
````

## Exercise 2.10

Given a geometric sequence with first term $1$ and common ratio $r$. Find the sum of the first 100 terms of the sequence if:
- $r=2$
- $r=\frac{1}{2}$
- $r=-3$
- $r=-\frac{5}{6}$
- $r=4.1$
- $r=0.7$

What can you conclude about the relationship between the common ratio of the geometric sequence with the sum of the first $n$ terms of the geometric sequence as $n$ grows larger and larger? i.e., as $n\rightarrow \infty$. You might want to try with more values of $r$.

````python
#YOUR CODE HERE
# Exercise 2.10: Geometric Sequence Analysis
# Calculate sum of first 100 terms for different common ratios
ratios = [2, 0.5, -3, -5/6, 4.1, 0.7]

for r in ratios:
    first_term = 1
    n = 100
    if r == 1:
        sum_n = first_term * n
    else:
        sum_n = first_term * (r ** n - 1) / (r - 1)
    print(f"r = {r}: Sum of first 100 terms = {sum_n}")

print("\nConclusion:")
print("When |r| < 1, the sum converges to a finite value as n approaches infinity.")
print("When |r| > 1, the sum grows without bound (diverges) as n approaches infinity.")
print("When |r| = 1 and r != 1 (i.e., r = -1), the sum oscillates.")
````

## Exercise 2.11 2019/NJC Tutorial/APGP/QB8 H2 Mathematics (Modified)

An athlete is considering two plans for a training program.

Plan I: Run 2000 m on day 1, 2400 m on day 2, and on each successive day increase the distance run by 400 m.

Plan II: Run 2000 m on day 1, 2200 m on day 2, and on each successive day multiply the distance run by a constant factor $\frac{11}{10}$.

- **(a)** For Plan I, find the least value of $n$ for which the total distance run on days 1 to $n$, inclusive, exceeds 200 000 m.

- **(b)** For Plan II, find the least value of N for which the distance run on day $N$ exceeds 16 000.

- **(c)** Assume that athlete $A$ decided to go with Plan I and another athlete $B$ decided to with Plan II, when will athlete $B$'s total distance run first exceeds athlete $A$'s total distance run?

````python
#YOUR CODE HERE
# Exercise 2.11: Training Program Comparison

# Part (a): Plan I - Find least n where total distance exceeds 200,000 m
# Arithmetic sequence: a = 2000, d = 400
# Sum formula: S_n = n/2 * (2a + (n-1)d)
n = 1
while True:
    total_distance = n / 2 * (2 * 2000 + (n - 1) * 400)
    if total_distance > 200000:
        print(f"Part (a): The least value of n for Plan I is {n}")
        print(f"Total distance on day {n}: {total_distance} m\n")
        break
    n += 1

# Part (b): Plan II - Find least N where distance on day N exceeds 16,000 m
# Geometric sequence: a = 2000, r = 11/10
# nth term: a_n = a * r^(n-1)
N = 1
while True:
    distance = 2000 * (11/10) ** (N - 1)
    if distance > 16000:
        print(f"Part (b): The least value of N for Plan II is {N}")
        print(f"Distance on day {N}: {distance} m\n")
        break
    N += 1

# Part (c): When does athlete B's total exceed athlete A's total?
day = 1
while True:
    # Plan I total (arithmetic)
    total_A = day / 2 * (2 * 2000 + (day - 1) * 400)
    
    # Plan II total (geometric): S_n = a(r^n - 1)/(r - 1)
    r = 11/10
    total_B = 2000 * (r ** day - 1) / (r - 1)
    
    if total_B > total_A:
        print(f"Part (c): Athlete B's total first exceeds Athlete A's total on day {day}")
        print(f"Athlete A's total: {total_A} m")
        print(f"Athlete B's total: {total_B} m")
        break
    day += 1
````
